# Read data

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import folium
```

```python
data = gpd.read_file('/content/CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_District_Level.shp')
```

```python
data
```

```python
data.plot(figsize=(12, 20))
plt.show()
```

```python
data.crs
```

```python
data = data.to_crs(epsg=4326)
```

```python
data.head()
```

```python
data.plot(figsize=(12, 20))
plt.show()
```

```python
map = folium.Map(location=[40.70, -73.94], zoom_start=10, tiles='CartoDB positron')
```

```python
for _, r in data.iterrows():
    # Without simplifying the representation of each borough,
    # the map might not be displayed
    sim_geo = gpd.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                           style_function=lambda x: {'fillColor': 'orange'})
    folium.Popup(r['Dist_Name']).add_to(geo_j)
    geo_j.add_to(map)
```

```python
map
```

```python
# Project to NAD83 projected crs
data = data.to_crs(epsg=2263)

# Access the centroid attribute of each polygon
data['centroid'] = data.centroid
```

```python
data.head()
```

```python
# Project to WGS84 geographic crs

# geometry (active) column
data = data.to_crs(epsg=4326)

# Centroid column
data['centroid'] = data['centroid'].to_crs(epsg=4326)
```

```python
for _, r in data.iterrows():
    lat = r['centroid'].y
    lon = r['centroid'].x
    folium.Marker(location=[lat, lon],
                  popup='length: {} <br> area: {}'.format(r['Shape_Leng'], r['Shape_Area'])).add_to(map)
```
